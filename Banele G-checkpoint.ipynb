```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Load data set
df = pd.read_csv("HEX.CSV")
df.head()
```

```python
#scatterplot Matrix
sns.pairplot(df, hue='variety')
plt.suptitle("Pair plot", y=1.02)
plt.show()
```

```python
sns.relplot(x= 'sepal.length', y='petal.length', hue='variety',data=df)
sns.relplot(x= 'petal.length', y='petal.width', hue='variety',data=df)
sns.relplot(x= 'sepal.length', y='sepal.width', hue='variety',data=df)
plt.show()
```

```python
#Visualizations
df.hist(figsize=(10,6), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=14)
plt.show()
```

```python
#Boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Boxplots Features')
plt.show()
```

```python
#Boxplots(Check for Outliers)
plt.figure(figsize=(12, 6))
for i, column in enumerate(df.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x="variety", y=column, data=df)
    plt.title(f"{column} by Variety")
    plt.tight_layout()
    plt.show()
```

```python
#Density plots
plt.figure(figsize=(8,4))
df.plot(kind='density', subplots=True, layout=(2,2),sharex=False)
plt.tight_layout()
plt.show()
```
